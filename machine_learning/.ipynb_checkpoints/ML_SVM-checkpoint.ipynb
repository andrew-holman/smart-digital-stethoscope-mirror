```python
def preprocess_data(data_sets):
    print('Starting to process entries...')
    training_data = []
    hasGone = 0
    for data_set in data_sets:
        averages, peak_widths, wavelengths, std_dev = preprocess_row3(data_set)
        training_data.append([averages, peak_widths, wavelengths, std_dev])
        if (hasGone == 0):
            hasGone = 1
    print('Complete')
    return training_data

def preprocess_row(data_set):
    data_set = abs(data_set)
    peak_buffer = numpy.array([])
    valley_counter = 0
    REQUIRED_PEAK_DIFF = 3
    REQUIRED_VALLEY_DIFF = 100
    PERCENTAGE_BASE_SAMPLES = .8
    TO_BEAT_BASE_VAL = 4
    
#     Gets lower area of array to get a baseline that spikes will need to beat
    base = numpy.average(numpy.sort(data_set)[0: math.floor(len(data_set) * .8)])
    
    std_dev = numpy.std(data_set) * 3
    averages = numpy.array([])
    peak_widths = numpy.array([])
    wavelengths = numpy.array([])
    for entry in data_set:
        if (entry > base * TO_BEAT_BASE_VAL):
            if (len(peak_buffer) == 0):
                if (valley_counter < REQUIRED_VALLEY_DIFF):
                    continue
                wavelengths = numpy.append(wavelengths, valley_counter)
                valley_counter = 0
            peak_buffer = numpy.append(peak_buffer, entry)
        elif(len(peak_buffer) != 0):
            if (len(peak_buffer) < REQUIRED_PEAK_DIFF):
                continue
            averages = numpy.append(averages, numpy.mean(peak_buffer))
            peak_widths = numpy.append(peak_widths, len(peak_buffer))
            peak_buffer = numpy.array([])
        else:
            valley_counter+=1
    
    return averages, peak_widths, wavelengths, std_dev

def preprocess_row2(data_set):
    data_set = abs(data_set)
    peak_buffer = numpy.array([])
    valley_counter = 0
    REQUIRED_PEAK_DIFF = 3
    REQUIRED_VALLEY_DIFF = 150
    PERCENTAGE_BASE_SAMPLES = .8
    TO_BEAT_BASE_VAL = 3
    DATA_BUFFER_SIZE = 5
    data_buffer = numpy.zeros(DATA_BUFFER_SIZE)
    
#     Gets lower area of array to get a baseline that spikes will need to beat
    base = numpy.average(numpy.sort(data_set)[0: math.floor(len(data_set) * .9)])
#     print(base * TO_BEAT_BASE_VAL + .1)
    top_ten = numpy.average(numpy.sort(data_set)[math.floor(len(data_set) * .97):])
    base = (top_ten / TO_BEAT_BASE_VAL + base)
    
    std_dev = numpy.std(data_set) * 3
    averages = numpy.array([])
    peak_widths = numpy.array([])
    wavelengths = numpy.array([])
    for entry in data_set:
        data_buffer = numpy.delete(data_buffer, 0)
        data_buffer = numpy.append(data_buffer, entry)
        if (numpy.mean(data_buffer) > base):
            if (len(peak_buffer) == 0):
                if (valley_counter < REQUIRED_VALLEY_DIFF):
                    continue
                wavelengths = numpy.append(wavelengths, valley_counter)
                valley_counter = 0
            peak_buffer = numpy.append(peak_buffer, entry)
        elif(len(peak_buffer) != 0):
            if (len(peak_buffer) < REQUIRED_PEAK_DIFF):
                continue
            averages = numpy.append(averages, numpy.mean(peak_buffer))
            peak_widths = numpy.append(peak_widths, len(peak_buffer))
            peak_buffer = numpy.array([])
        else:
            valley_counter+=1
    
    return averages, peak_widths, wavelengths, std_dev

def preprocess_row3(data_set):
    data_set = abs(data_set)
    peak_buffer = numpy.array([])
    valley_counter = 0
    REQUIRED_PEAK_DIFF = 5
    REQUIRED_VALLEY_DIFF = 100
    PERCENTAGE_BASE_SAMPLES = .85
    TO_BEAT_BASE_VAL = 2
    DATA_BUFFER_SIZE = 15
    data_buffer = numpy.zeros(DATA_BUFFER_SIZE)
    
#     Gets lower area of array to get a baseline that spikes will need to beat
    base = numpy.average(numpy.sort(data_set)[0: math.floor(len(data_set) * PERCENTAGE_BASE_SAMPLES)])
#     print(base * TO_BEAT_BASE_VAL + .1)
    top_ten = numpy.average(numpy.sort(data_set)[math.floor(len(data_set) * PERCENTAGE_BASE_SAMPLES):])
    base = (top_ten / TO_BEAT_BASE_VAL + base)
    
    std_dev = numpy.std(numpy.sort(data_set)[0: math.floor(len(data_set) * PERCENTAGE_BASE_SAMPLES)])
    averages = numpy.array([])
    peak_widths = numpy.array([])
    wavelengths = numpy.array([])
    for entry in data_set:
        data_buffer = numpy.delete(data_buffer, 0)
        data_buffer = numpy.append(data_buffer, entry)
        if (numpy.mean(data_buffer) > base): # If we are currently in a peak
            if (len(peak_buffer) == 0): # First time in this?
                if (valley_counter < REQUIRED_VALLEY_DIFF):
                    continue
                wavelengths = numpy.append(wavelengths, valley_counter)
                valley_counter = 0
            peak_buffer = numpy.append(peak_buffer, entry)
        elif(len(peak_buffer) != 0):
            if (len(peak_buffer) < REQUIRED_PEAK_DIFF):
                continue
            averages = numpy.append(averages, numpy.mean(peak_buffer))
            peak_widths = numpy.append(peak_widths, len(peak_buffer))
            peak_buffer = numpy.array([])
        else:
            valley_counter+=1
    
    return averages, peak_widths, wavelengths, std_dev
```
